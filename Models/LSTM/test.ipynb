```python
import numpy as np
```

```python
import json
with open("data.json", "r", encoding="utf-8") as file:
    data = json.load(file)  # Now 'data' is properly defined
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


# Prepare dataset
X_train, y_train = [], []
labels = {}
for i, intent in enumerate(data["intents"]):
    for pattern in intent["patterns"]:
        X_train.append(pattern)
        y_train.append(i)
    labels[i] = intent["tag"]

# Tokenize and pad sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_train_padded = pad_sequences(X_train_seq, padding='post')

# Convert labels to categorical
y_train = tf.keras.utils.to_categorical(y_train, len(labels))

# Build LSTM model
model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=16),
    tf.keras.layers.LSTM(16),
    tf.keras.layers.Dense(len(labels), activation="softmax")
])

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(X_train_padded, y_train, epochs=50)

while True:
    def get_lstm_response(user_input):
        if user_input == "bye":
            print("Byee...")
            exit()
        seq = tokenizer.texts_to_sequences([user_input])
        padded = pad_sequences(seq, maxlen=X_train_padded.shape[1], padding='post')
        prediction = model.predict(padded)
        tag_index = np.argmax(prediction)
        return labels[tag_index]
    # man()

# Example usage

    val = input("Enter your Query:")
    print(get_lstm_response(val))  # Output: Future Bound Tech


```
